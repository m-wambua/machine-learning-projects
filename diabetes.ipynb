```python
global_mean = df_train_full.outcome.mean()
for col in features:
    df_group = df_train_full.groupby(by=col).outcome.agg(['mean'])
    df_group['diff'] = df_group['mean'] - global_mean
    df_group['risk'] = df_group['mean'] / global_mean
    display(df_group)
```

```python
from sklearn.metrics import mutual_info_score
```

```python
def calculate_mi(series):
    return mutual_info_score(series, df_train_full.outcome)

df_mi = df_train_full[features].apply(calculate_mi)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='MI')


display(df_mi.head())
display(df_mi.tail())
```

```python
from sklearn.feature_extraction import DictVectorizer

train_dict=df_train[features].to_dict(orient='records')
```

```python
train_dict[0]
```

```python
dv=DictVectorizer(sparse=False)
dv.fit(train_dict)
```

```python
X_train=dv.transform(train_dict)
```

```python
X_train.shape
```

```python
dv.get_feature_names()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X_train,y_train)
```

```python
val_dict=df_val[features].to_dict(orient='records')
X_val=dv.transform(val_dict)
```

```python
lin_reg.predict(X_val)
```

```python
y_pred=lin_reg.predict(X_val)
```

```python
outcome=y_pred>0.5
```

```python
lin_reg_predict_accuracy=(y_val==outcome).mean()

##The accuracy while using linear regression
```

```python
## Stochastic Gradient Descent
from sklearn.linear_model import SGDRegressor
sgd_reg=SGDRegressor(max_iter=1000,tol=1e-3,penalty=None,eta0=0.1)
sgd_reg.fit(X_train,y_train.ravel())
```

```python
y_pred_stochastic=sgd_reg.predict(X_val)
```

```python
outcome_stochastic=y_pred_stochastic>0.5
```

```python
stochastic_decent_acc=(y_val==outcome_stochastic).mean()
```
